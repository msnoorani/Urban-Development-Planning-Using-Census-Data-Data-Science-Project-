```python
#my_mockcensus_copy[(my_mockcensus_copy['Surname'] == 'Riley') & (my_mockcensus_copy['Street'] == 'White Branch')]
```

```python
# Defining clean-age for age

def clean_age(age):
    # Convert to string first to handle all cases
    str_age = str(age).strip().lower()

    # Keep empty strings as-is (don't convert to NaN)
    if str_age == '':
        return age  # Return original value

    # Handle word numbers
    if str_age == 'twenty eight':
        return 28

    # Clean string - keep only digits, minus, and single decimal point
    cleaned = []
    decimal_count = 0
    for char in str_age:
        if char.isdigit():
            cleaned.append(char)
        elif char == '-' and not cleaned:  # Only allow minus at start
            cleaned.append(char)
        elif char == '.' and decimal_count == 0:
            cleaned.append(char)
            decimal_count += 1

    cleaned_str = ''.join(cleaned)

    # Convert to number
    try:
        num = float(cleaned_str)
        if -0.0001 < num < 120.0001:  # Allow slight floating point tolerance
            age_int = int(round(num))
            # Convert 0 to 1
            return 1 if age_int == 0 else age_int
    except (ValueError, TypeError):
        pass

    return np.nan  # This still converts invalid entries to NaN

# Apply cleaning directly to Age column
my_mockcensus_copy['Age'] = my_mockcensus_copy['Age'].apply(clean_age)

# Final verification
print("\nAge distribution:")
print(my_mockcensus_copy['Age'].describe())
print(f"\nTotal rows: {len(my_mockcensus_copy)}")
print("\nUnique values in Age column after cleaning:")
print(my_mockcensus_copy['Age'].unique())
```

```python
# Get unique age values after cleaning and dropping NaN
unique_ages = my_mockcensus_copy['Age'].unique()

# Print the results
print("Unique Age values after cleaning:")
print(np.sort(unique_ages))  # Sorted for easier reading

# Additional information
print(f"\nNumber of unique ages: {len(unique_ages)}")
print(f"Age range: {unique_ages.min()} to {unique_ages.max()}")
```

```python
# Grouping individuals acccording to their residences

def residence(my_mockcensus_copy, number, street):
    new = my_mockcensus_copy.groupby([number,street]).size().reset_index(name = 'count')
    new['Residence'] = new['count'].astype(int)
    new.loc[new['count'] == 1, 'Residence'] = 1
    address_ = new.set_index([number, street])['Residence'].to_dict()
    return my_mockcensus_copy.apply(lambda x: address_.get((x[number],x[street])),axis=1)

my_mockcensus_copy['Residence'] = residence(my_mockcensus_copy, 'House Number', 'Street')
```

```python
# Assuming old age starts from 66;

Elderly_class = len(my_mockcensus_copy[(my_mockcensus_copy['Age'] >= 66) & (my_mockcensus_copy['Age'] <= 110)])

# Future deaths using crude death rate to estimate 10 years

Elderly_deaths = ((len(my_mockcensus_copy)/1000) * Crude_Death_Rate) * 10

Future_elderly = Elderly_class - Elderly_deaths

print(f'The estimated number of future elderly people is {round(Future_elderly)}')

Percentage_elderly = ((Future_elderly/len(my_mockcensus_copy))*100)

print(f'{Percentage_elderly:.2f}% make up the estimated percentage of elderly people')
```
